```python
import pandas
import codecs

path = 'map.html'
template = codecs.open(path, 'r', 'utf-8');
contents = template.read()
template.close

import os

path = 'TextFiles'
os_info = os.walk(path)
counter = 0
for roots, dirs, files in os_info:
    if(counter is 0):
        coord_dir = files
    counter = 1
    

for file in coord_dir:
    coord_file = open('TextFiles/' + file, 'r', encoding = 'utf-8', errors = 'ignore')
    coords = coord_file.read()
    coord_file.close()
    lines = contents.splitlines()
    first = '\n'.join(lines[0:103])
    last = '\n'.join(lines[103:])
   
    for line in coords.splitlines():
        first += '\t\t\t' + line + '\n'
    html_string = first + last
    file_name = file.replace('.txt', '')
    template = open('Maps/' + file_name + '_map.html', 'w+')
    template.write(html_string)
    template.close()
   
```
